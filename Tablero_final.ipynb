```python
import plotly.express as px

def mapa_pyb(df, cohorte):

    df_filtered = df[(df["cohorte"] == cohorte)]
    df_filtered = df_filtered[df_filtered['porcentaje'] > 0]
    filtered_df1 = df_filtered.replace({
        'Desertores precoces': 'Precoces',
        'Desertores tempranos': 'Tempranos',
        'Desertores al corte': 'Al corte',
        'Desertores después del corte': 'Despues del corte'
    })

    # Definir el color_map con los nuevos nombres
    color_map = {
        'Precoces': "#5BB084",
        'Tempranos': "#EBBF70",
        'Al corte': "#5CE1E6",
        'Despues del corte': "#D2A6F1"
    }

    # Crear el gráfico
    fig = px.scatter_mapbox(
        filtered_df1,
        lat="latitud",
        lon="longitud",
        mapbox_style="carto-positron",
        zoom=5,
        hover_name="municipio",
        color="porcentaje_desertor",
        color_discrete_map=color_map,
        size="porcentaje",
        size_max=50,
        labels={"porcentaje_desertor": "Tipo de desertor"},
        hover_data={'porcentaje_desertor': True, 'porcentaje': True, "latitud": False, "longitud": False}
    )

    # Configurar el layout
    fig.update_layout(
        font=dict(
            family="sans-serif Open Sans",
            size=10
        ),
        margin={"r":80, "t":20, "l":80, "b":20},
        legend=dict(
            title=dict(
                text="Tipo de desertor",
                font=dict(size=10)
            ),
        )
    )

    return fig
```

*Choropleth*

```python
import plotly.graph_objects as go

def mapa_ch(df,cohorte,tipo_desertor,geojson_data):
    df_filtered = df[(df["cohorte"] == cohorte) & (df["porcentaje_desertor"] == tipo_desertor)]
    df_filtered = df_filtered[df_filtered['porcentaje'] > 0]

    # Excluir el valor máximo
    max_value = df_filtered["porcentaje"].max()
    df_excluded_max = df_filtered[df_filtered['porcentaje'] != max_value]

    #figura para el valor máximo
    fig_max = px.choropleth_mapbox(
        df_filtered[df_filtered['porcentaje'] == max_value],
        geojson=geojson_data,
        locations="clave_combinada",
        featureidkey='properties.clave_combinada',
        color="porcentaje",
        color_continuous_scale="bluered",  # Cambiar a tu color preferido
        mapbox_style="carto-positron",
        zoom=4.7,
        center={"lat":4.570860,"lon":-74.2973328},
        hover_name='municipio',
        hover_data={'departamento': True, 'porcentaje': True,"clave_combinada":False},
        # labels={'porcentaje_desertor':'Porcentaje desertores'},
    )

    #figura para los valores restantes
    fig_rest = px.choropleth_mapbox(
        df_excluded_max,
        geojson=geojson_data,
        locations="clave_combinada",
        featureidkey='properties.clave_combinada',
        color="porcentaje",
        color_continuous_scale="bluyl",
        mapbox_style="carto-positron",
        zoom=5,
        center={"lat":4.570860,"lon":-74.2973328},
        hover_name='municipio',
        hover_data={'departamento': True, 'porcentaje': True,"clave_combinada":False},
        labels={'porcentaje_desertor':'Porcentaje desertores'},
        range_color=[df_excluded_max["porcentaje"].min(), df_excluded_max["porcentaje"].max()]
    )

    # Combinar las dos figuras
    fig_max.update_traces(marker=dict(opacity=1))
    fig_rest.add_trace(fig_max.data[0])

    fig_rest.update_layout(
        font=dict(
            family="sans-serif Open Sans",
            size=12
        ),
        margin={"r":20,"t":20,"l":20,"b":20},
        coloraxis_colorbar=dict(
            title="Porcentaje desertores",
            titleside="right",
            titlefont=dict(size=14),
            tickfont=dict(size=12),
            len=0.8  #longitud de la barra de colores
        )
    )

    return fig_rest
```

**Spider estrato, porcentaje y condición estudiantil**

```python
#Función
def spider_plot(df, cohorte):
    #Filtrar los datos según la cohorte seleccionada
    data1 = df[(df['semestre_inicia_programa'] == cohorte) & (df['tipo'] == 'foraneo')]
    data2 = df[(df['semestre_inicia_programa'] == cohorte) & (df['tipo'] == 'no_foraneo')]
    categories = ['Estrato 1', 'Estrato 2', 'Estrato 3', 'Estrato 4', 'Estrato 5', 'Estrato 6']
    rf = []
    rnf = []
    labels_rf = []
    labels_rnf = []

    for i in range(1, 7):
        # Filtrar los datos para el estrato actual en 'data1' y 'data2', pues foraneo no tiene estrato 6 en la tabla
        fd1 = data1[data1['estrato'] == i]['porcentaje_desertores']
        fd2 = data2[data2['estrato'] == i]['porcentaje_desertores']

        #Verificar si hay datos para el estrato actual en 'data1'
        if not fd1.empty:
            rf.append(float(fd1.values[0]))
            labels_rf.append(f'Estrato {i}: {float(fd1.values[0]):.2f}%')
        else:
            rf.append(0)
            labels_rf.append(f'Estrato {i}: 0%')

        #Verificar si hay datos para el estrato actual en 'data2'
        if not fd2.empty:
            rnf.append(float(fd2.values[0]))
            labels_rnf.append(f'Estrato {i}: {float(fd2.values[0]):.2f}%')
        else:
            rnf.append(0)
            labels_rnf.append(f'Estrato {i}: 0%')

    fig = go.Figure()

    fig.add_trace(go.Scatterpolar(
        name='Foráneo',
        r=rf,
        theta=categories,
        fill='toself',
        line=dict(color='#5F9EA0'),
        text=labels_rf,
        hoverinfo='text'
    ))

    fig.add_trace(go.Scatterpolar(
        name='No foráneo', r=rnf,
        theta=categories,
        fill='toself',
        line=dict(color='#F08080'),
        text=labels_rnf,
        hoverinfo='text'
    ))

    fig.update_layout(
        polar=dict(
            radialaxis=dict(
                visible=True,
                range=[0, 35],
            )),
        showlegend=True
    )

    return fig
```
